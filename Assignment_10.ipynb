```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
iris_df = pd.DataFrame(data, columns=iris.feature_names)
iris_df['target'] = target
```

```python
num_clusters = 3
```

```python
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(data)
```

```python
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
iris_df['cluster'] = labels
```

```python
plt.figure(figsize=(10, 6))
```

```python
for cluster_num in range(num_clusters):
    cluster_data = iris_df[iris_df['cluster'] == cluster_num]
    plt.scatter(cluster_data.iloc[:, 0], cluster_data.iloc[:, 1], label=f'Cluster {cluster_num}')
```

```python
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=200, color='red', label='Cluster Centers')
```
